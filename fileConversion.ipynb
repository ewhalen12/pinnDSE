# Saving geometry in an alternative format

```python
import numpy as np
import pandas as pd
import pyvista as pv
import deepxde as dde
import pickle

from pinnDSE.geom import *
from pinnDSE.viz import *
from pinnDSE.elasticity import *
from pinnDSE.bc import *
```

```python
# load mesh and solver results using pyNastran
op2File = 'data/controlArm/v1.0/controlArm01_nom.op2'
geom = MeshGeom(op2File, thickness=0.25)
plotScalarFields(geom.mesh, geom.resDf)
```

```python
# write the geometry object as a mesh file and pickle file so that it can be loaded without pyNastran
geom.saveInAltFormat()
```

```python
# reload using the alternative format
meshFile = 'data/controlArm/v1.0/controlArm01_nom.vtk'
pickleFile = 'data/controlArm/v1.0/controlArm01_nom.p'
geom2 = MeshGeom(meshFile=meshFile, pickleFile=pickleFile)
plotScalarFields(geom2.mesh, geom2.resDf)
```
