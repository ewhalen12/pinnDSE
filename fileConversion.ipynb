# Saving geometry in an alternative format

```python
import numpy as np
import pandas as pd
import pyvista as pv
import deepxde as dde
import pickle

from pinnDSE.geom import *
from pinnDSE.viz import *
from pinnDSE.elasticity import *
from pinnDSE.bc import *
```

```python
# load mesh and solver results using pyNastran
op2File = 'data/controlArm/v1.0/controlArm01_nom.op2'
geom = MeshGeom(op2File, thickness=0.25)
plotScalarFields(geom.mesh, geom.resDf)
```

```python
# write the geometry object as a mesh file and pickle file so that it can be loaded without pyNastran
geom.saveInAltFormat()
```

```python
# reload using the alternative format
meshFile = 'data/controlArm/v1.0/controlArm01_nom.vtk'
pickleFile = 'data/controlArm/v1.0/controlArm01_nom.p'
geom2 = MeshGeom(meshFile=meshFile, pickleFile=pickleFile)
plotScalarFields(geom2.mesh, geom2.resDf)
```

```python
# load mesh and solver results using pyNastran
op2File = 'data/connector/v1.0/connectorDesign03.op2'
geomCon = MeshGeom(op2File, thickness=6.35)
plotScalarFields(geomCon.mesh, geomCon.resDf)
```

```python
# write the geometry object as a mesh file and pickle file so that it can be loaded without pyNastran
geomCon.saveInAltFormat()
```

```python
# reload using the alternative format
meshFile = 'data/connector/v1.0/connectorDesign03.vtk'
pickleFile = 'data/connector/v1.0/connectorDesign03.p'
geonCon2 = MeshGeom(meshFile=meshFile, pickleFile=pickleFile)
plotScalarFields(geonCon2.mesh, geonCon2.resDf)
```

```python
op2File = 'data/cantilever/v1.0/cantilever01.op2'
geomB = MeshGeom(op2File, thickness=0.05, center=False, scale=False)

plotScalarFields(geomB.mesh, geomB.resDf, shape=(1,5), size=200)
```

```python
geomB.saveInAltFormat2()
```

```python
# reload using the alternative format
pickleFile = 'data/cantilever/v1.0/cantilever01.p'
geomB2 = MeshGeom(pickleFile=pickleFile, altFormat=2)
plotScalarFields(geomB2.mesh, geomB2.resDf, shape=(1,5), size=200)
```

```python
plotter = pv.Plotter(border=False, notebook=True)
plotter.set_background('white')
plotter.add_mesh(geomB2.mesh, show_edges=False)
plotter.show(window_size=(600,600), cpos='xy');
```
